### Theory

In last couple of chapters, we saw some corner detectors like Harris etc. They are rotation-invariant, which means, even if the image is rotated, we can find the same corners. It is obvious because corners remain corners in rotated image also. But what about scaling? A corner may not be a corner if the image is scaled. 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('../images/berlin.jpg')
#Next line is just for arranging the figure size
#Image size is extended to highlight the keypoints
plt.figure(figsize=(20,20))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Berlin')
plt.show()
```

### So now let's see SIFT functionalities available in OpenCV. 

Let's start with keypoint detection and draw them. First we have to construct a SIFT object. We can pass different parameters to it which are optional.

sift.detect() function finds the keypoint in the images. You can pass a mask if you want to search only a part of image. Each keypoint is a special structure which has many attributes like its (x,y) coordinates, size of the meaningful neighbourhood, angle which specifies its orientation, response that specifies strength of keypoints etc.



```python
#Converting our pixels in to gray values
gray= cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

#sift
sift = cv2.SIFT_create()
kp = sift.detect(gray,None)

#Drawing key points
img=cv2.drawKeypoints(gray,kp,image)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Berlin with key points')
plt.show()
```

```python
img = cv2.drawKeypoints(image,kp,img,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.figure(figsize=(20,20))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Berlin with highlighted key points')
plt.show()
```
