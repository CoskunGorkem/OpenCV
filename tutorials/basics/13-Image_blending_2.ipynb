```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('../images/python_logo.png')
img2 = cv2.imread('../images/opencv_logo.png')
```

### If the image shapes are same, images can be blended together. 
To do that, we need to resize both of them by using cv2.resize method

```python
width = 2000 #that will be the final width of the images
height = 2000 #that will be the final height of the images
dim = (width, height)
 
# resize both of the images
resized_img1 = cv2.resize(img1, dim, interpolation = cv2.INTER_AREA)
resized_img2 = cv2.resize(img2, dim, interpolation = cv2.INTER_AREA)

print('Original Dimensions of img1 : ',img1.shape)
print('Original Dimensions of img2 : ',img2.shape)
print("==============================================")
print('Resized Dimensions of img1: ',resized_img1.shape)
print('Resized Dimensions of img2: ',resized_img2.shape)
```

```python
#lets blend them together by arranging their weights on the final image
img1_weight = 0.5
img2_weight = 0.5
dst = cv2.addWeighted(resized_img1,img1_weight,resized_img2,img2_weight,0)
```

```python
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.title('Blended')
plt.show()
```
