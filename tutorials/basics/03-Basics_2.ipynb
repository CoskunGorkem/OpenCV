```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('../images/dog.jpg')
```

# Cropping

```python
#Finding size of the image
height, width = image.shape[:2]

#Top  left of cropped rectangle
start_height, start_width = int(height * .25), int(width * .25)

#Bottom right of cropeed rectangle
end_height, end_width = int(height * .75), int(width * .75)

#Simply use indexing to crop out the rectangle we desire
cropped = image[start_height:end_height , start_width:end_width]
```

```python
row, col = 1, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
 
axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')
cv2.imwrite('original_image.png', image)

axs[1].imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
axs[1].set_title('Cropped Image')
cv2.imwrite('cropped_image.png', cropped)

plt.show()
```

# Put features in image


```python
# Making a sqare
cv2.rectangle(image, (10, 10), (50, 50), 255, -2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Square Image')

plt.show()
```

```python
cv2.putText(image, 'Hi There!', 
            (36,170), cv2.FONT_HERSHEY_COMPLEX, 0.7, (255,0,0), 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Texted Image')
plt.show()
```

```python
cv2.line(image, (0,200), (200,200), (0,255,0), 5)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('What about lines?')
plt.show()
```

```python
cv2.circle(image, (80, 75), 20, (0,0,255), 5) 
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('What about circles?')
plt.show()
```
